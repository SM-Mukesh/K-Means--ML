## Model Building

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
```

```python
model.fit(df.drop('Private',axis=1))
```

```python
model.cluster_centers_
```

Since we had the Target for this dataset ,we defined K as 2

for educational purpose lets build a model with number of k values and check the elbow method

```python
k = range(1,8)
a = []

for i in k:
    model = KMeans(n_clusters=i)
    model.fit(df.drop('Private',axis=1))
    a.append(model.inertia_)
    
plt.plot(k,a,marker='o')
plt.grid()
```
